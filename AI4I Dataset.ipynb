# Data Loading & Exploration {#data-loading}

Loading the AI4I 2020 Predictive Maintenance dataset and performing initial exploratory analysis to understand the data structure and characteristics.

````python
# Data Preprocessing and Feature Engineering
print("Performing data preprocessing and feature engineering...")

# Define feature columns (actual column names from the dataset)
numeric_features = ['Air temperature', 'Process temperature', 
                   'Rotational speed', 'Torque', 'Tool wear']

# Create processed feature set
X_processed = X[numeric_features].copy()

# Create binary target for machine failure prediction
y_binary = y['Machine failure'].copy()

# Encode categorical Type variable
le = LabelEncoder()
X_processed['Type_encoded'] = le.fit_transform(X['Type'])

print(f"Original features shape: {X.shape}")
print(f"Processed features shape: {X_processed.shape}")
print(f"Target shape: {y_binary.shape}")

# Display feature statistics
print("\nFeature Statistics:")
print(X_processed.describe())

# Check for outliers using IQR method
print("\nOutlier Analysis:")
outlier_summary = {}
for col in numeric_features:
    Q1 = X_processed[col].quantile(0.25)
    Q3 = X_processed[col].quantile(0.75)
    IQR = Q3 - Q1
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR
    
    outliers = X_processed[(X_processed[col] < lower_bound) | (X_processed[col] > upper_bound)]
    outlier_count = len(outliers)
    outlier_percentage = (outlier_count / len(X_processed)) * 100
    
    outlier_summary[col] = {
        'count': outlier_count,
        'percentage': outlier_percentage
    }
    
    print(f"{col}: {outlier_count} outliers ({outlier_percentage:.2f}%)")

# Target distribution analysis
print(f"\nTarget Distribution:")
print(f"No Failure: {(y_binary == 0).sum()} ({((y_binary == 0).sum() / len(y_binary)) * 100:.2f}%)")
print(f"Failure: {(y_binary == 1).sum()} ({((y_binary == 1).sum() / len(y_binary)) * 100:.2f}%)")

failure_rate = y_binary.mean()
print(f"Overall failure rate: {failure_rate:.4f} ({failure_rate * 100:.2f}%)")

print("Data preprocessing completed")
````

## Deep Learning Implementation

Advanced neural network approaches for predictive maintenance including Multi-Layer Perceptron (MLP) classification and autoencoder-based anomaly detection to enhance prediction accuracy and operational reliability.

````python
# Multi-Layer Perceptron (MLP) Model Definition
print("Defining Multi-Layer Perceptron (MLP) architecture...")

# Neural Network for Classification
mlp_model = Sequential([
    Dense(128, activation='relu', input_shape=(X_train_dl.shape[1],)),
    Dropout(0.3),
    Dense(64, activation='relu'),
    Dropout(0.3),
    Dense(32, activation='relu'),
    Dropout(0.2),
    Dense(1, activation='sigmoid')
])

mlp_model.compile(
    optimizer='adam',
    loss='binary_crossentropy',
    metrics=['accuracy']
)

print("MLP model architecture defined successfully")
print(f"Model parameters: {mlp_model.count_params():,}")
````

````python
# MLP Model Training
print("Training Multi-Layer Perceptron model...")

# Early stopping to prevent overfitting
early_stopping = EarlyStopping(
    monitor='val_loss', 
    patience=10, 
    restore_best_weights=True
)

# Train the model
history = mlp_model.fit(
    X_train_dl, y_train_dl,
    epochs=100,
    batch_size=32,
    validation_split=0.2,
    callbacks=[early_stopping],
    verbose=0
)

print(f"Training completed - Final epoch: {len(history.history['loss'])}")
print("MLP model training finished successfully")
````

````python
# MLP Model Evaluation
print("Evaluating MLP model performance...")

# Make predictions
mlp_predictions_prob = mlp_model.predict(X_test_dl, verbose=0)
mlp_predictions = (mlp_predictions_prob > 0.5).astype(int).flatten()

# Calculate metrics
mlp_accuracy = accuracy_score(y_test_dl, mlp_predictions)
mlp_f1 = f1_score(y_test_dl, mlp_predictions)

print(f"MLP Performance:")
print(f"  Accuracy: {mlp_accuracy:.4f}")
print(f"  F1-Score: {mlp_f1:.4f}")
print("MLP evaluation completed")
````

````python
# Autoencoder Model Definition
print("Defining Autoencoder architecture for anomaly detection...")

# Use only normal samples for training autoencoder
normal_mask = y_train_dl == 0
X_normal = X_train_dl[normal_mask]

# Define autoencoder architecture
input_dim = X_train_dl.shape[1]
encoding_dim = 16

input_layer = Input(shape=(input_dim,))
encoder = Dense(64, activation='relu')(input_layer)
encoder = Dense(32, activation='relu')(encoder)
encoder = Dense(encoding_dim, activation='relu')(encoder)

decoder = Dense(32, activation='relu')(encoder)
decoder = Dense(64, activation='relu')(decoder)
decoder = Dense(input_dim, activation='linear')(decoder)

autoencoder = Model(input_layer, decoder)
autoencoder.compile(optimizer='adam', loss='mse')

print(f"Autoencoder architecture defined successfully")
print(f"Normal samples for training: {len(X_normal)}")
print(f"Encoding dimension: {encoding_dim}")
````

````python
# Autoencoder Model Training
print("Training Autoencoder model...")

# Train autoencoder on normal samples only
autoencoder_history = autoencoder.fit(
    X_normal, X_normal,
    epochs=100,
    batch_size=32,
    validation_split=0.2,
    callbacks=[early_stopping],
    verbose=0
)

print(f"Autoencoder training completed - Final epoch: {len(autoencoder_history.history['loss'])}")
print("Autoencoder model training finished successfully")
````

````python
# Autoencoder Anomaly Detection
print("Performing anomaly detection with Autoencoder...")

# Calculate reconstruction errors
reconstructed = autoencoder.predict(X_test_dl, verbose=0)
reconstruction_errors = np.mean(np.square(X_test_dl - reconstructed), axis=1)

# Determine threshold (95th percentile of training reconstruction errors)
normal_errors = np.mean(np.square(X_normal - autoencoder.predict(X_normal, verbose=0)), axis=1)
threshold = np.percentile(normal_errors, 95)

# Classify anomalies
anomaly_predictions = (reconstruction_errors > threshold).astype(int)

# Calculate metrics for autoencoder
ae_accuracy = accuracy_score(y_test_dl, anomaly_predictions)
ae_f1 = f1_score(y_test_dl, anomaly_predictions)

print(f"Autoencoder Performance:")
print(f"  Threshold: {threshold:.4f}")
print(f"  Accuracy: {ae_accuracy:.4f}")
print(f"  F1-Score: {ae_f1:.4f}")
print("Autoencoder anomaly detection completed")
````

````python
# Cost-Benefit Analysis for Predictive Maintenance
print("Performing cost-benefit analysis for maintenance decisions...")

# Define cost parameters (in USD) - these would be company-specific
COST_PARAMETERS = {
    'preventive_maintenance': 1500,      # Cost of scheduled maintenance
    'failure_downtime': 25000,           # Cost of unplanned downtime
    'emergency_repair': 8000,            # Cost of emergency repairs
    'false_alarm': 500,                  # Cost of unnecessary maintenance
    'production_hour': 2000,             # Value of one production hour
    'avg_downtime_hours': 8              # Average hours of downtime per failure
}

def calculate_maintenance_cost(tp, fp, tn, fn, costs):
    """Calculate total cost based on confusion matrix and cost parameters"""
    
    # True Positives: Correctly predicted failures - preventive maintenance
    cost_tp = tp * costs['preventive_maintenance']
    
    # False Positives: False alarms - unnecessary maintenance
    cost_fp = fp * costs['false_alarm']
    
    # True Negatives: Correctly predicted no failure - no cost
    cost_tn = 0
    
    # False Negatives: Missed failures - expensive!
    cost_fn = fn * (costs['failure_downtime'] + costs['emergency_repair'])
    
    total_cost = cost_tp + cost_fp + cost_tn + cost_fn
    
    return {
        'preventive_maintenance_cost': cost_tp,
        'false_alarm_cost': cost_fp,
        'missed_failure_cost': cost_fn,
        'total_cost': total_cost
    }

# Calculate costs for each model
print("Calculating costs for each model...")
cost_analysis = []

for name, model_data in all_models.items():
    if 'predictions' in model_data:
        # Calculate confusion matrix
        cm = confusion_matrix(y_test, model_data['predictions'])
        tn, fp, fn, tp = cm.ravel()
        
        # Calculate costs
        costs = calculate_maintenance_cost(tp, fp, tn, fn, COST_PARAMETERS)
        
        cost_analysis.append({
            'Model': name,
            'True_Positives': tp,
            'False_Positives': fp,
            'False_Negatives': fn,
            'Preventive_Maintenance_Cost': costs['preventive_maintenance_cost'],
            'False_Alarm_Cost': costs['false_alarm_cost'],
            'Missed_Failure_Cost': costs['missed_failure_cost'],
            'Total_Cost': costs['total_cost']
        })

# Create cost analysis dataframe
cost_df = pd.DataFrame(cost_analysis)
cost_df = cost_df.sort_values('Total_Cost')

print("\nCost-Benefit Analysis Results:")
print("=" * 200)
print(cost_df.to_string(index=False))

# Calculate cost savings vs baseline
baseline_cost = cost_df['Total_Cost'].max()  # Worst performing model
best_cost = cost_df['Total_Cost'].min()      # Best performing model
savings = baseline_cost - best_cost

print(f"\nCost Analysis Summary:")
print(f"Highest cost model: {cost_df.iloc[-1]['Model']} - ${cost_df.iloc[-1]['Total_Cost']:,.0f}")
print(f"Lowest cost model: {cost_df.iloc[0]['Model']} - ${cost_df.iloc[0]['Total_Cost']:,.0f}")
print(f"Potential annual savings: ${savings:,.0f}")

# Visualize cost breakdown
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(16, 8))

# Cost comparison
models_for_plot = cost_df['Model'].tolist()
total_costs = cost_df['Total_Cost'].tolist()

bars = ax1.bar(models_for_plot, total_costs, color=['#2E86AB', '#A23B72', '#F18F01', '#C73E1D', '#4CAF50', '#FF9800'][:len(models_for_plot)])
ax1.set_title('Total Cost Comparison by Model', fontsize=14, fontweight='bold')
ax1.set_ylabel('Total Cost (USD)')
ax1.tick_params(axis='x', rotation=45)

# Add value labels on bars
for bar, cost in zip(bars, total_costs):
    ax1.text(bar.get_x() + bar.get_width()/2, bar.get_height() + max(total_costs)*0.01,
             f'${cost:,.0f}', ha='center', va='bottom', fontweight='bold')

# Cost breakdown for best model
best_model_row = cost_df.iloc[0]
cost_categories = ['Preventive\nMaintenance', 'False\nAlarms', 'Missed\nFailures']
cost_values = [
    best_model_row['Preventive_Maintenance_Cost'],
    best_model_row['False_Alarm_Cost'],
    best_model_row['Missed_Failure_Cost']
]

bars2 = ax2.bar(cost_categories, cost_values, color=['#4CAF50', '#FF9800', '#F44336'])
ax2.set_title(f'Cost Breakdown - {best_model_row["Model"]}', fontsize=14, fontweight='bold')
ax2.set_ylabel('Cost (USD)')

# Add value labels
for bar, cost in zip(bars2, cost_values):
    ax2.text(bar.get_x() + bar.get_width()/2, bar.get_height() + max(cost_values)*0.01,
             f'${cost:,.0f}', ha='center', va='bottom', fontweight='bold')

plt.tight_layout()
plt.show()

# ROI calculation
print(f"\nReturn on Investment (ROI) Analysis:")
print(f"Annual cost savings from best model: ${savings:,.0f}")
print(f"Implementation cost (estimated): $50,000")
print(f"ROI in first year: {((savings - 50000) / 50000) * 100:.1f}%")
print(f"Payback period: {50000 / savings * 12:.1f} months")

print("Cost-benefit analysis completed")
````

````python
# Advanced Neural Network Architectures for Predictive Maintenance
print("Implementing advanced neural network architectures...")

# Check if TensorFlow is available and proceed accordingly
if TENSORFLOW_AVAILABLE:
    print("✓ Using TensorFlow/Keras for deep learning models")
    
    # Import additional layers needed
    from tensorflow.keras.layers import Add, Activation, BatchNormalization, Input, LSTM, Attention, Layer

    # 1. Residual Neural Network (ResNet-style) for Industrial Data
    print("\\n1. Building Residual Neural Network")

    def create_residual_block(x, units, dropout_rate=0.3):
        """Create a residual block for better gradient flow"""
        # Main path
        main = Dense(units, activation='relu')(x)
        main = BatchNormalization()(main)
        main = Dropout(dropout_rate)(main)
        main = Dense(units, activation='relu')(main)
        main = BatchNormalization()(main)
        
        # Skip connection
        if x.shape[-1] != units:
            skip = Dense(units, activation='linear')(x)
        else:
            skip = x
        
        # Add skip connection
        return Add()([main, skip])

    # Build ResNet model
    input_layer = Input(shape=(X_train_scaled.shape[1],))
    x = Dense(128, activation='relu')(input_layer)
    x = BatchNormalization()(x)
    
    # Residual blocks
    x = create_residual_block(x, 64)
    x = create_residual_block(x, 32)
    x = create_residual_block(x, 16)
    
    # Output layer
    output_layer = Dense(1, activation='sigmoid')(x)
    
    resnet_model = Model(inputs=input_layer, outputs=output_layer)
    resnet_model.compile(optimizer=Adam(learning_rate=0.001), 
                        loss='binary_crossentropy', 
                        metrics=['accuracy', 'precision', 'recall'])
    
    print(f"ResNet Model created with {resnet_model.count_params()} parameters")

    # 2. Multi-Layer Perceptron with Professional Architecture
    print("\\n2. Building Professional MLP Model")
    
    mlp_model = Sequential([
        Dense(256, activation='relu', input_shape=(X_train_scaled.shape[1],)),
        BatchNormalization(),
        Dropout(0.4),
        
        Dense(128, activation='relu'),
        BatchNormalization(),
        Dropout(0.3),
        
        Dense(64, activation='relu'),
        BatchNormalization(),
        Dropout(0.2),
        
        Dense(32, activation='relu'),
        Dropout(0.2),
        
        Dense(1, activation='sigmoid')
    ])
    
    mlp_model.compile(optimizer=Adam(learning_rate=0.001),
                     loss='binary_crossentropy',
                     metrics=['accuracy', 'precision', 'recall'])
    
    print(f"MLP Model created with {mlp_model.count_params()} parameters")
    
    # 3. Autoencoder for Anomaly Detection
    print("\\n3. Building Autoencoder for Anomaly Detection")
    
    # Encoder
    encoder_input = Input(shape=(X_train_scaled.shape[1],))
    encoded = Dense(32, activation='relu')(encoder_input)
    encoded = Dense(16, activation='relu')(encoded)
    encoded = Dense(8, activation='relu')(encoded)
    
    # Decoder
    decoded = Dense(16, activation='relu')(encoded)
    decoded = Dense(32, activation='relu')(decoded)
    decoded = Dense(X_train_scaled.shape[1], activation='sigmoid')(decoded)
    
    autoencoder = Model(encoder_input, decoded)
    autoencoder.compile(optimizer=Adam(learning_rate=0.001), loss='mse')
    
    print(f"Autoencoder created with {autoencoder.count_params()} parameters")
    
    # Store models for later use
    deep_learning_models = {
        'ResNet': resnet_model,
        'MLP': mlp_model,
        'Autoencoder': autoencoder
    }
    
    print("✓ TensorFlow-based neural networks ready for training")

else:
    print("WARNING:  TensorFlow not available. Using sklearn neural network alternatives...")
    
    # Alternative approach using sklearn
    from sklearn.neural_network import MLPClassifier
    from sklearn.ensemble import IsolationForest
    
    # 1. Multi-Layer Perceptron using sklearn
    print("\\n1. Building sklearn MLP Classifier")
    
    mlp_sklearn = MLPClassifier(
        hidden_layer_sizes=(256, 128, 64, 32),
        activation='relu',
        solver='adam',
        alpha=0.001,
        learning_rate='adaptive',
        learning_rate_init=0.001,
        max_iter=500,
        early_stopping=True,
        validation_fraction=0.2,
        n_iter_no_change=10,
        random_state=42
    )
    
    print("✓ sklearn MLP Classifier configured")
    
    # 2. Isolation Forest for Anomaly Detection (alternative to autoencoder)
    print("\\n2. Building Isolation Forest for Anomaly Detection")
    
    isolation_forest = IsolationForest(
        contamination=0.1,  # Expected proportion of outliers
        random_state=42,
        n_estimators=100
    )
    
    print("✓ Isolation Forest configured")
    
    # 3. Simple deep-like network using sklearn
    print("\\n3. Building Advanced sklearn Neural Network")
    
    advanced_mlp = MLPClassifier(
        hidden_layer_sizes=(512, 256, 128, 64, 32, 16),
        activation='relu',
        solver='adam',
        alpha=0.0001,
        learning_rate='adaptive',
        learning_rate_init=0.001,
        max_iter=1000,
        early_stopping=True,
        validation_fraction=0.2,
        n_iter_no_change=15,
        random_state=42
    )
    
    print("✓ Advanced sklearn Neural Network configured")
    
    # Store models for later use
    deep_learning_models = {
        'MLP_sklearn': mlp_sklearn,
        'Advanced_MLP': advanced_mlp,
        'Isolation_Forest': isolation_forest
    }
    
    print("✓ sklearn-based neural network alternatives ready for training")

print(f"\\nTotal models prepared: {len(deep_learning_models)}")
print("Ready to proceed with model training and evaluation")
````

````python
# 🧪 A/B TESTING FRAMEWORK DEMONSTRATION
print("🧪 A/B TESTING FRAMEWORK DEMONSTRATION")
print("=" * 60)

# Simplified A/B Testing Class (notebook version)
import hashlib
import random
import numpy as np
from scipy import stats

class SimpleABTest:
    def __init__(self, name, control_model, treatment_model, traffic_split=0.5):
        self.name = name
        self.control_model = control_model
        self.treatment_model = treatment_model
        self.traffic_split = traffic_split
        self.assignments = {}
        self.results = {'control': [], 'treatment': []}
    
    def assign_traffic(self, user_id):
        """Assign user to control or treatment group"""
        if user_id in self.assignments:
            return self.assignments[user_id]
        
        # Consistent assignment based on user_id hash
        hash_value = int(hashlib.md5(str(user_id).encode()).hexdigest(), 16)
        assignment = 'control' if (hash_value % 100) / 100 < self.traffic_split else 'treatment'
        self.assignments[user_id] = assignment
        return assignment
    
    def get_prediction(self, features, assignment):
        """Get prediction from appropriate model"""
        if assignment == 'control':
            return self.control_model.predict(features)
        else:
            return self.treatment_model.predict(features)
    
    def record_result(self, user_id, prediction, actual):
        """Record test result"""
        assignment = self.assignments.get(user_id)
        if assignment:
            self.results[assignment].append((prediction, actual))

# Setup test data
print("Setting up A/B Test Experiment...")
print("Using Random Forest as control model...")
print("Using Balanced Random Forest as treatment model...")

# Use the models we have available
control_model = rf_model  # Random Forest from previous cell
treatment_model = brf_model  # Balanced Random Forest

print(f"Control model accuracy: {control_model.score(X_processed, y_binary):.4f}")
print(f"Treatment model accuracy: {treatment_model.score(X_processed, y_binary):.4f}")

# Initialize A/B test
ab_test = SimpleABTest(
    name="ML_Model_Comparison_Test",
    control_model=control_model,
    treatment_model=treatment_model,
    traffic_split=0.5
)

print(f"\nCHART: A/B Test Configuration:")
print(f"Test Name: {ab_test.name}")
print(f"Traffic Split: {ab_test.traffic_split * 100}% control, {(1-ab_test.traffic_split) * 100}% treatment")

# Simulate some test data for demonstration
test_indices = np.random.choice(len(X_processed), size=200, replace=False)
X_test_sample = X_processed.iloc[test_indices]
y_test_sample = y_binary.iloc[test_indices]

print(f"\n🔬 Running A/B Test with {len(X_test_sample)} samples...")

# Track predictions for both models
control_results = []
treatment_results = []

for i, (idx, row) in enumerate(X_test_sample.iterrows()):
    user_id = f"user_{i}"
    # Assign traffic
    assignment = ab_test.assign_traffic(user_id)
    
    # Get prediction based on assignment
    pred = ab_test.get_prediction(row.values.reshape(1, -1), assignment)
    actual = y_test_sample.iloc[i]
    
    # Record result
    ab_test.record_result(user_id, pred[0], actual)
    
    if assignment == 'control':
        control_results.append((pred[0], actual))
    else:
        treatment_results.append((pred[0], actual))

# Calculate metrics
def calculate_accuracy(results):
    if not results:
        return 0
    correct = sum(1 for pred, actual in results if pred == actual)
    return correct / len(results)

control_accuracy = calculate_accuracy(control_results)
treatment_accuracy = calculate_accuracy(treatment_results)

print(f"\nTREND: A/B Test Results:")
print(f"Control Group (Random Forest):")
print(f"  - Samples: {len(control_results)}")
print(f"  - Accuracy: {control_accuracy:.4f}")

print(f"Treatment Group (Balanced Random Forest):")
print(f"  - Samples: {len(treatment_results)}")
print(f"  - Accuracy: {treatment_accuracy:.4f}")

# Statistical significance test
if len(control_results) > 10 and len(treatment_results) > 10:
    control_correct = [1 if pred == actual else 0 for pred, actual in control_results]
    treatment_correct = [1 if pred == actual else 0 for pred, actual in treatment_results]
    
    # Two-sample t-test
    t_stat, p_value = stats.ttest_ind(control_correct, treatment_correct)
    
    print(f"\nCHART: Statistical Analysis:")
    print(f"T-statistic: {t_stat:.4f}")
    print(f"P-value: {p_value:.4f}")
    
    if p_value < 0.05:
        winner = "Treatment" if treatment_accuracy > control_accuracy else "Control"
        print(f"TARGET: Result: {winner} model shows statistically significant improvement!")
    else:
        print("CHART: Result: No statistically significant difference detected")

# Business impact simulation
print(f"\nCOST: Business Impact Analysis:")
failure_cost = 10000  # Cost per undetected failure
maintenance_cost = 500  # Cost per unnecessary maintenance

# Simulate business metrics
def calculate_business_impact(results):
    false_negatives = sum(1 for pred, actual in results if pred == 0 and actual == 1)
    false_positives = sum(1 for pred, actual in results if pred == 1 and actual == 0)
    
    cost = (false_negatives * failure_cost) + (false_positives * maintenance_cost)
    return cost, false_negatives, false_positives

control_cost, control_fn, control_fp = calculate_business_impact(control_results)
treatment_cost, treatment_fn, treatment_fp = calculate_business_impact(treatment_results)

print(f"Control Model (Random Forest):")
print(f"  - False Negatives: {control_fn} (missed failures)")
print(f"  - False Positives: {control_fp} (unnecessary maintenance)")
print(f"  - Total Cost: ${control_cost:,}")

print(f"Treatment Model (Balanced Random Forest):")
print(f"  - False Negatives: {treatment_fn} (missed failures)")
print(f"  - False Positives: {treatment_fp} (unnecessary maintenance)")
print(f"  - Total Cost: ${treatment_cost:,}")

cost_savings = control_cost - treatment_cost
print(f"\nINSIGHT: Potential Savings: ${cost_savings:,}")

print(f"\n✅ A/B Testing Framework successfully demonstrated!")
print("This framework enables:")
print("• Controlled model rollouts with traffic splitting")
print("• Statistical significance testing")
print("• Business impact measurement")
print("• Data-driven deployment decisions")
print("• Risk mitigation through gradual model updates")
````
